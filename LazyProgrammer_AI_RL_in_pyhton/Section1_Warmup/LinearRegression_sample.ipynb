# generate point to simulate a linear regression

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sn
import random
```

```python
np.random.seed(0)
n = 25
x = np.linspace(0,10,n) + np.random.randn(n)/5
y = np.random.randn(n)/10
```

```python
plt.scatter(x,y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
plt.figure(figsize = (10,10))
plt.scatter(X_train, y_train)
plt.scatter(X_test, y_test)
```

```python
model = LinearRegression()
hist = model.fit(X=X_train.reshape(-1,1),y=y_train)
```

```python
X_predict_input = np.linspace(0, 10, 100).reshape(-1,1)
y_predict = model.predict(X_predict_input)
```

```python
plt.figure(figsize = (10,10))
plt.scatter(X_train, y_train)
plt.scatter(X_test, y_test)
plt.plot(X_predict_input, y_predict)
plt.xlim=(0,11)
```
