```python
# define the gridworld  class
class WindyGridWorld():
    
    def __init__(self, rows, columns, start_position):
        self.rows = rows
        self.columns = columns
        #self.all_states = [(i,j) for i in range(rows) for j in range(columns)]
        self.i = start_position[0]
        self.j = start_position[1]
        
    def set_rewards_actions(self, rewards, actions, probs):
        self.rewards = rewards
        self.actions = actions
        self.probs = probs
        self.all_states = set(self.actions.keys()) | set(self.rewards.keys())
        #print (self.all_states)
    
    def set_state(self, s):
        self.i = s[0]
        self.j = s[1]
    
    def current_state(self):
        return self.i,self.j
    
    def reset(self):
        self.i = 2
        self.j = 0
        return self.current_state()
    
    def undo_move(self, action):
        if action in self.actions[(self.i,self.j)]:
            if action == 'U':
                self.i += 1
            elif action == 'R':
                self.j -= 1
            elif action == 'L':
                self.j += 1
            else:
                self.i -= 1
        # should never happen
        assert (self.current_state() in self.all_states)
 
    def move(self, action):
        cur_state = (self.i, self.j)
        a = action
        next_action_prob = self.probs[(cur_state,a)]
        next_actions = list(next_action_prob.keys())
        next_probs = list(next_action_prob.values())
        next_state_idx = np.random.choice(len(next_actions), p=next_probs)
        self.i = next_actions[next_state_idx][0]
        self.j = next_actions[next_state_idx][1]
        return self.rewards.get((self.i,self.j),0)

    def is_terminal (self, s):
        return s not in self.actions
    
    def game_over(self):
        return (self.i,self.j) in self.actions
```

```python
# define the gridworld  class
class GridWorld():
    
    def __init__(self, rows, columns, start_position):
        self.rows = rows
        self.columns = columns
        #self.all_states = [(i,j) for i in range(rows) for j in range(columns)]
        self.i = start_position[0]
        self.j = start_position[1]
        
    def set_rewards_actions(self, rewards, actions):
        self.rewards = rewards
        self.actions = actions
        self.all_states = set(self.actions.keys()) | set(self.rewards.keys())
        print (self.all_states)
    
    def set_state(self, s):
        self.i = s[0]
        self.j = s[1]
    
    def current_state(self):
        return self.i,self.j

    def reset(self):
        self.i = 2
        self.j = 0
        return self.current_state()
    
    def get_next_state(self, s, a):
        i, j = s[0], s[1]
        #print(f"s : ({i},{j})")
        #print(a)
        #print(self.actions[(i,j)])
        if a in self.actions[(i,j)]:
            if a == 'U':
                i -= 1
            elif a == 'R':
                j += 1
            elif a == 'L':
                j -= 1
            else:
                i += 1
        #print(f"s2 : ({i},{j})")
        return i,j
    
    def undo_move(self, action):
        if action in self.actions[(self.i,self.j)]:
            if action == 'U':
                self.i += 1
            elif action == 'R':
                self.j -= 1
            elif action == 'L':
                self.j += 1
            else:
                self.i -= 1
        # should never happen
        assert (self.current_state() in self.all_states)
 
    def move(self, action):
        if action in self.actions[(self.i,self.j)]:
            if action == 'U':
                self.i -= 1
            elif action == 'R':
                self.j += 1
            elif action == 'L':
                self.j -= 1
            else:
                self.i += 1
        return self.rewards.get((self.i,self.j),0)

    def is_terminal (self, s):
        return s not in self.actions
    
    def game_over(self):
        return (self.i,self.j) in self.actions
```

```python
def print_values(V,g):
    for i in range(g.rows):
        print("---------------------------")
        for j in range(g.columns):
            v = V.get((i,j),0)
            if v >= 0:
                print(" %.2f|" % v, end="")
            else:
                print("%.2f|" % v, end="")
        print ("")

def print_policy(P,g):
    for i in range(g.rows):
        print("---------------------------")
        for j in range(g.columns):
            a = P.get((i,j),' ')
            print(" %s |" % a, end="")
        print ("")

ACTION_SPACE = ('U', 'D', 'L', 'R')

def init_windy_grid_world_penalized(step_cost, start):
    grid = WindyGridWorld(3,4,start)
    rewards = {
        (2,0):step_cost,
        (1,0):step_cost,
        (0,0):step_cost,
        (2,1):step_cost,
        (0,1):step_cost,
        (2,2):step_cost,
        (1,2):step_cost,
        (0,2):step_cost,
        (2,3):step_cost,
        (0,3):1,
        (1,3):-1
    }
    grid.set_rewards_actions(rewards, actions, probs)
    return grid

def max_dict(d):
    '''
    return the argmax and max value from a dictionnary
    '''
    max_val = max(d.values())
    max_keys = [key for key, val in d.items() if val == max_val]
    return np.random.choice(max_keys), max_val

def epsilon_greedy(Q, state, eps=0.1):
    '''
    return an actions based on the epsilon greedy concept : exploration vs exploitation
    random action if < epx else best action
    '''
    p = np.random.random()
    if p < eps:
        a = np.random.choice(ACTION_SPACE)
    else:
        a = Q[state]
    return a
    
```

```python
SMALL_ENOUGH = 1e-3


def play_game(grid, policy, max_iteration=20, step_cost=0):
    
    #start state
    s = grid.reset()
    
    #choose action
    a = epsilon_greedy(policy, s)
    
    cur_game_states = [s]
    cur_game_actions = [a]
    cur_game_rewards = [0]
    
    # until game over or reach a max of game iterations store the path of the current game
    for _ in range(max_iteration):
        r = grid.move(a)
        s = grid.current_state()
        
        cur_game_rewards.append(r)
        cur_game_states.append(s)
        
        if not grid.game_over():
            break
        else:
            a = epsilon_greedy(policy,s)
            cur_game_actions.append(a)
            
    return cur_game_states, cur_game_actions, cur_game_rewards


def main(step_cost, tot_iteration):
    
    gamma = 0.9    
        
    # initialize G and returns
    #g = init_grid_world_penalized(step_cost, (2,0))
    g = GridWorld(3,4,(2,0))
    g.set_rewards_actions(rewards=rewards, actions=actions)
    
    print(f"rewards : ")
    print_values(g.rewards,g)
    
    # initialize a random policy
    policy = {}
    for s in g.actions.keys():
        policy[s] = np.random.choice(ACTION_SPACE)
    
    # initialize Q(s,a) and returns
    Q = {}
    sample_counts = {}
    state_sample_count = {}
    states = g.all_states
    for s in states:
        if s in g.actions: #not a terminal space
            Q[s]={}
            sample_counts[s]={} # returns
            state_sample_count[s]=0 
            for a in ACTION_SPACE:
                Q[s][a]=0
                sample_counts[s][a]=0
        else: # terminal state
            pass
   
    # repeat until convergence
    deltas = []
    #loop on tot_iteration
    for t in range(tot_iteration):
        if t % 1000 == 0:
            print(t)
      
        # play one game based on a random policy
        biggest_change = 0
        print_policy(policy,g)
        cur_game_states, cur_game_actions, cur_game_rewards = play_game(g, policy)
    
        #create a list of only state-action
        states_actions = list(zip(cur_game_states,cur_game_actions))
        
        #print(states_actions)
        #compute G and V based on results
        
        G = 0
        game_length = len(cur_game_states)
       
        for i in range(game_length-2,-1,-1): 
            last_state = cur_game_states[i]
            #print(last_state)
            last_action = cur_game_actions[i]
            G = cur_game_rewards[i+1]+gamma*G
            

            if (last_state, last_action) not in states_actions.pop():
                q_old = Q[last_state][last_action]
                sample_counts[last_state][last_action] += 1
                lr = 1 / sample_counts[last_state][last_action]
                Q[last_state][last_action] = q_old + lr * (G - q_old)
                
                policy[last_state] = max_dict(Q[last_state])[0]
                
                # update state sample count
                state_sample_count[last_state] +=1
                
                #update delta
                biggest_change = max(biggest_change, np.abs(q_old - Q[last_state][last_action]))
            deltas.append(biggest_change)

    plt.plot(deltas)
    plt.show()
    
    print("finale policy")
    print_policy(policy,g)
    
    # find V
    V = {}
    for s, Qs in Q.items():
        V[s] = max_dict(Q[s])[1]
    
    print("final values:")
    print_values(V,g)
    
    print("state_sample_count: ")
    state_sample_count_arr = np.zeros((g.rows, g.columns))
    for i in range(g.rows):
        for j in range(g.columns):
            if (i,j) in state_sample_count:
                state_sample_count_arr[i,j] = state_sample_count[(i,j)]
    df = pd.DataFrame(state_sample_count_arr)
    print(df)
```
